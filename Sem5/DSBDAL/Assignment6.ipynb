```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
#histogram for numeric feature
numeric_features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
df[numeric_features].hist(bins=25,figsize=(15,10),edgecolor = 'blue')
plt.suptitle('Histogram of Iris feature')
plt.show()

#Analysis
#SepalLength: value lies between 4.3 to 7.9,most values 5.0-6.5, right-skewed
#SepalWidth: value between 2.0-4.4,most values 2.8-3.4, right-skewed
#PetalLength: values between 1-7,clustered 1-2, 3-5, 5-7, not strongly skewed
#PetalWidth: values between 0.1-2.5, clusteres 0.1-0.5, 1.0-1.5,1.8-2.5
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df[numeric_features])
plt.title('Box plot')
plt.show()

# Feature      Median     Range     Outlier  
#SepalLength    5.8       4.3-7.9    No strong 
#SepalWidth     3         2.1-4      2,4.1,4.2
#PetalLength    4.1       1.1-6.9    No strong
#PetalWidth     1.3       0.3-2.1    No strong

#IQR more ->distributed/spread out ->more variability->strong difference in species
#IQR small ->closed together->
```

```python
#Box plots of features with respect to species like iris-setosa etc.
plt.figure(figsize=(10, 6))

# Box plot for SepalLengthCm

sns.boxplot(x="Species", y="SepalLengthCm", data=df)
plt.title("SepalLengthCm")
```
