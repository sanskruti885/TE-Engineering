```python
# Import required libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer

# Download necessary datasets (run only once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
# Load the given text files
with open("PRINCIPALS_MESSAGE.txt", "r", encoding="utf-8") as f:
    doc1 = f.read()

with open("DIRECTORS_MESSAGE.txt", "r", encoding="utf-8") as f:
    doc2 = f.read()

# Combine documents into a list
documents = [doc1, doc2]

print("Documents Loaded Successfully!")
```

```python
print("Document 1 Sample:\n", doc1[:200])
print("\nDocument 2 Sample:\n", doc2[:200])
```

```python
#tokenization break text into smaller part
nltk.download('punkt_tab')
token_doc1 = word_tokenize(doc1);
token_doc2 = word_tokenize(doc2);

print('\nDoc1 tokens:')
print(token_doc1[:40])
print('\nDoc2 tokens:')
print(token_doc2[:40])
```

```python
#assign grammtical label like noun, preposition,verb
nltk.download('averaged_perceptron_tagger_eng')
pos_tags1 = pos_tag(token_doc1)
pos_tags2 = pos_tag(token_doc2)

print("\nPOS Tags1:")
print(pos_tags1[:50])

print("\nPOS Tags2:")
print(pos_tags2[:50])
```

```python
#stop word removal words like in, a, the ,as not required
stop_words = set(stopwords.words('english'))
new_token1 = [word for word in token_doc1 if word.lower() not in stop_words]
new_token2 = [word for word in token_doc2 if word.lower() not in stop_words]

print("\nAfter Stop Word Removal1:")
print(new_token1[:50])

print("\nAfter Stop Word Removal2:")
print(new_token2[:50])
```

```python
#stemming reduce to original or root form
#removes suffix like ing, ed, ly, is
stemmer = PorterStemmer()

stemmed_word1 = [stemmer.stem(word) for word in new_token1]
stenned_word2 = [stemmer.stem(word) for word in new_token2]

print("\nStemmed Words1:")
print(stemmed_word1[:50])

print("\nStemmed Words2:")
print(stenned_word2[:50])
```

```python
#lemmatization means reduce word to base or dictionary form
#better good, rocks rock
# Lemmatization
lemmatizer = WordNetLemmatizer()

lemmatized_words1 = [lemmatizer.lemmatize(word) for word in new_token1]
lemmatized_words2 = [lemmatizer.lemmatize(word) for word in new_token2]

print("\nLemmatized Words1:")
print(lemmatized_words1[:50])

print("\nLemmatized Words2:")
print(lemmatized_words2[:50])
```

```python
#TF-IDF term frequency, inverse document frequency
#TF : How many times a word occurs
#IDF : how unique a word is
#TF(t,d)=Total number of terms in document d/ Number of times term t appears in document d​
#IDF(t,d)=log(N/DF(t))

documents = [doc1, doc2]

vectorizer = TfidfVectorizer(stop_words='english')

tfidf_matrix = vectorizer.fit_transform(documents)

feature_names = vectorizer.get_feature_names_out()
```

```python
import numpy as np
np.set_printoptions(threshold=np.inf)
print("\nFeatures Words: \n", feature_names)
print("\nTF-IDF Matrix: \n", tfidf_matrix.toarray())
```

```python
import pandas as pd

df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df
```

```python
# Show top 10 words as a table
for i in range(len(df)):
    print(f"\nTop words in Document {i+1}:\n")
    display(df.iloc[i].sort_values(ascending=False).head(10).to_frame(name="TF-IDF Score"))
```
