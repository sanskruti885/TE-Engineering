```python
import numpy as np
import pandas as py
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0]) #since not numberic need to take mode()[0]
df['deck'] = df['deck'].fillna(df['deck'].mode()[0])
```

```python
df['sex']= df['sex'].astype('string')
df['embarked']=df['embarked'].astype('string')
df['who']=df['who'].astype('string')
df['alive']=df['alive'].astype('string')
df['who']=df['who'].astype('string')
```

```python
#map sex and age
sns.boxplot(x='sex',y='age',data=df)
```

```python
#how many survived from male and female or different age
plt.figure(figsize=(10,6)) #width and height
sns.boxplot(x='sex',y='age',hue='survived',data=df)
```
